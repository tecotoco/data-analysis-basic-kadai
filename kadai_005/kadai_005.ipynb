```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# 0から10までの幅で配列を等間隔に100個作成する
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(y1, linewidth=2.0)

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
y3 = np.random .normal(15, 5, 2000)
# ヒストグラムをプロット
axes[0, 2].hist(y3)

# データを生成する
np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))
axes[1, 0].scatter(x4, y4)

plt.show()
```
